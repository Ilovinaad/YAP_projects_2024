# Проектная работа
# Cоздание презентации на основе дашбордов с данными о конференциях TED. 

**Описание проекта**

C помощью Tableau будет проведено иследование и визуальное сопровождение особенностей истории TED-конференций.

**Целью** исследования является выявление особенностей истории TED-конференций.

**Этапы** работы:

1. Подготовим и изучим данные файлов
2. Построим дашборд «История выступлений»
3. Создадим дашборд «Тематики выступлений»
4. Создадим дашборд «Авторы выступлений»
5. Построим дашборд на свободную тему - «Особенности проведение конференций TED в Великобритании»
6. Создадим презентацию, в которую войдут сделанные ранее дашборды.


## Шаг. Загрузка и изучение данных
- Загрузили данные в Tableau.
- С помощью Union объединили файлы с данными выступлений в единую таблицу. Объединили таблицу с данными выступлений, справочники конференций и авторов с помощью Relationship.
- Построили визуализации распределения конференций во времени.

***Выявили год, с которого начинается рост количества конференций - 2001. Резкий скачкообразный рост начался в 2008 г, с 2017 г следует снижение количества конференций TED.***

- Настроили фильтр, который исключит все выступления до найденного порогового года на уровне источника данных.

## Шаг. Создание дашборда «История выступлений»

- Создали лист «Выступления по странам», оформили визуализацию типа map, которая показала процент выступлений в каждой из стран.
- Создали лист «Выступления по годам» с визуализацией типа area charts, использовали измерение Main Tag для разбивки выступлений по категориям (для этого создали сет, в который вошли топ-N основных тематик выступлений).
- Создали новый лист «Процент выступлений по тематикам» и отобразили процент наблюдений по категориям за каждый год.
- Задали фильтр по полю Film Date и применили ко всем визуализациям на основе текущего набора данных.
- Создали дашборд «История выступлений» по макету и Настроили Action на дашборде так, чтобы при выборе страны на листе «Выступления по странам» фильтр применялся и к остальным листам.

***Выводы и наблюдения по 2 Шагу:*** 

*1. Чаще всего конференции TED проходили в таких странах как: США, Канада, Великобритания и Австралия.*

*2. В топ-10 наиболее популярных категорий выступлений входят искусство, бизнес, культура, дизайн, образование, глобальные проблемы, наука, социальные изменения, общество, технологии.*

*3. В 2020г на конференциях выступления проходили со всеми выше перечисленными темами, однако в 2021г сократилось количество докладов и, в основном, звучали только темы технологии, общество и глобальные проблемы.*

**Изменение состава выступлений в 2020-2021 гг и количества конференций может быть связано с экономическими, политическими или социальными факторами, которые появились в мире во время возникновения коронавирусной инфекции.**

## Шаг. Создание дашборда «Тематики выступлений»

- Создали лист «Выступления по тематике» с круговой диаграммой по тематикам выступлений и отобразили на графике топ-10 категорий по количеству выступлений, а менее популярные объединили в категорию «Другие» (использовали для этого сет).
- Создали лист «Тематики и страны» с визуализацией типа highlight table с указанием процента выступлений на конкретную тему в каждой стране. Отсортировали полученную таблицу по убыванию числа просмотров.
- Создали три вычисляемых поля: 

Duration, min — длительность выступления в минутах;

Applause by Duration — отношение числа аплодисментов к длительности выступления в секундах;

Laughter by Duration — отношение числа ситуаций, когда аудитория смеялась, к длительности выступления в секундах.

- Создали лист «Аплодисменты по тематикам» с диаграммой размаха и добавили во всплывающую подсказку название, автора и значение Applause by Duration. Отсортировали категории на графике по убыванию медианного значения Applause by Duration.
- Создали новый лист «Смех по тематикам». На этой визуализации заменили Applause by Duration на Laughter by Duration.
- Создали два новых листа «Гистограмма длительности» и «Гистограмма просмотров». 
- Создали новый лист «Связь длительности с просмотрами» с визуализацией типа scatter plot.
- Создали дашборд «Тематики выступлений» по макету и настроили Action так, чтобы при выборе тематики на листе «Выступления по тематике» фильтр применялся к остальным визуализациям на дашборде, кроме таблицы «Тематики и страны».

***Выводы и наблюдения по 3 Шагу:*** 

*1. Самые популярные категории выступлений: искусство, бизнес, культура, дизайн, образование, глобальные проблемы, наука, социальные изменения, общество, технологии. Остальные категории были объединены под названием 'Other' ('Другие').*

*2. В топ-3 странах по количеству выступлений  темы конференций  в основном различаются. В Великобритании было проведено больше конференций на тему технологии, наука, глобальные проблемы, дизайн и бизнес, чем в Канаде.*

*3. Выступления на тему исскуства, образования и общество чаще вызывают аплодисменты аудитории, реже - технология и наука.   Выступлению 'Политический прогресс, совершенный женщинами - Что будет следующим? ' Сесилии Ричардс аплодировали больше остальных.*

*4. Чаще всего темы образования и общества вызывают смех у публики, а мешьше всего - глобальные проблемы и наука. Самым смешным лектором также стала Сесилия Ричардс.*

*5. самым просматриваемым выступлением  стало 'Язык вашего тела может описать, кто вы', а самое долгое выступление - '3 секрета успеха Netflix'. В основном, как таковой связи между длительностью выступления и количеством просмотров нет. Однако, большая часть конференций обладает менее, чем 20 млн просмотров.*


## Шаг. Создание дашборда «Авторы выступлений»

- Создали сет, который объединил наиболее популярные области деятельности авторов (Speaker Occupation) и настроили сет так, чтобы числом областей деятельности можно было управлять с помощью целочисленного параметра Top-N Speaker Occupations.
- Создали лист «Области деятельности авторов» с пузырьковой диаграммой (packed bubbles) и настроили визуализацию так, что она не показывала категорию «Другие».
- Создали вычисляемое поле Talks by Author, которое показало максимальное количество выступлений у автора (использовали LOD для фиксации измерения Author Id).
- Создали лист «Распределение числа выступлений» со столбчатой диаграммой.
- Создали лист «Авторы по числу выступлений», отсортировали таблицу по убыванию числа выступлений.
- Создали лист «Выступления и число просмотров», отсортировали таблицу по убыванию числа просмотров.
- Создали дашборд «Авторы выступлений» по макету и настроили Actions на дашборде так, чтобы при выборе области деятельности на визуализации «Области деятельности авторов» фильтровались таблицы «Авторы по числу выступлений» и «Выступления и число просмотров». При выборе автора в таблице «Авторы по числу выступлений» фильтровалась таблица «Выступления и число просмотров».

***Выводы и наблюдения по 4 Шагу:***

*1. Среди областей деятельности у авторов преобладают художник, автор, журналист и писатель, предприниматель и группа 'неопределено'.*

*2. На одного автора обычно приходится одно выступление. Ханс Рослинг выступал чаще всего - 8 раз.*

*3. Том Вуеч выступал чаще своих коллег в области дизайна - 4 раза. Меньше всего смотрели его выступление с названием 'Учимся использовать астролябию 13го века'.*

## Шаг. Создание дашборда на свободную тему «Особенности проведение конференций TED в Великобритании»

- Создали лист «Топ-3 стран по количеству проведенных в них TED-конференций» с визуализацией типа highlight table с указанием количества конференций по темам в таких странах, как США, Великобритания и Канада. Отсортировали полученную таблицу по убыванию числа конференций.
- Создали лист «Топ-10 категорий выступлений в Великобритании по количеству конференций» с круговой диаграммой по тематикам выступлений и отобразили на графике топ-10 категорий по количеству выступлений, а менее популярные объединили в категорию «Другие» (использовали для этого сет).
- Создали лист «Профессиональные области лекторов Великобритании» со столбчатой диаграммой и отсортировали по убыванию количества конференций.
- Создали лист «Связь количества аплодисментов и смеха на выступлениях лекторов в Великобритании» с визуализацией типа scatter plot.

***Выводы и наблюдения по 5 Шагу:***

*1. Великобритания занимает 2 место по количеству проведенных конференций по каждой из 10 самых многочисленных категорий.*

*2. Среди лекторов конференций TED в Великобритании преобладают люди с профессиональной областью журналист, писатель, дизайнер, художник и экономист. Эти люди в основном выступали по теме культуры, дизайна и технологии. Также заметно, что они могут выступать по темам, приближенным или даже невходящим в их профобластью.*

*3. Связи между аплодисменами и смехом в зале как таковой нет, однако, но большинство зрителей лекций смеются до 10 и аплодируют до 4 раз за одно выступление. За сатирическое выступление Рори Бремнера зрители засмеялись 55 раз! В среднем, за выступление британских авторов люди аплодировали 2 раза и смеялись 3 раза.*

## Общий вывод по проекту

Исходя из полученных нами данных о конференциях, можно сделать такие заключения:

1. В основном, англоговорящие страны США, Великобритания и Австралия, а также частично франкоговорящая Канада принимали самое большое количество конференций TED;

2. Пик всех конференций приходился на 2008-2012 гг, а среди основных тем были наука, технологии, дизайн и социальные проблемы.

3. Такие темы, как общество и искусство преобладают по количеству аплодисментов, меньше всего смех в зале вызывают темы технологии и социальные проблемы.

4. Во второй по количеству конференций стране (Великобритания) среди аторов выступлений преобладают журналисты и писатели. Также эти авторы могут читать лекции по темам, не входящих в их профессиональные области.
